```python
import os
import cv2
import numpy as np
import glob as gb
from matplotlib import pyplot as plt
from patchify import patchify
from PIL import Image
import segmentation_models as sm
from tensorflow.keras.metrics import MeanIoU
from keras.models import load_model
from sklearn.preprocessing import MinMaxScaler, StandardScaler
scaler = MinMaxScaler()
```

```python
root_directory = './semantic_drone_dataset/dataset'
IMG_WIDTH = 224
IMG_HEIGHT = 224
IMG_CHANNELS = 3
```

```python
train_dataset = [] 
files=gb.glob(root_directory+'/original_images/*.jpg')
for file in files :
    image = cv2.imread(file, 1)  #Read each image as BGR
    image = cv2.resize(image, (IMG_WIDTH, IMG_HEIGHT)) 
    image = np.array(image)             
    train_dataset.append(image)
print("original images loaded")
```

```python
mask_dataset = []  
files=gb.glob(root_directory+'/label_images_masks/*.png')
for file in files :
    mask = cv2.imread(file, 1)  #Read each image as BGR
    mask = cv2.cvtColor(mask,cv2.COLOR_BGR2RGB)
    image = cv2.resize(mask, (IMG_WIDTH, IMG_HEIGHT))  #Try not to resize for semantic segmentation                           
    mask_dataset.append(image)
print("mask images loaded")
```

```python
train = np.array(train_dataset)
mask_dataset =  np.array(mask_dataset)
print("shape origin => ", train.shape)
print("shape mask => ", mask_dataset.shape)
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Conv2D,MaxPooling2D,Conv2DTranspose,BatchNormalization,Concatenate,ReLU,LeakyReLU,Add,Activation,Input,MaxPool2D
from tensorflow.keras.layers import GlobalAveragePooling2D,AveragePooling2D,UpSampling2D
```

```python
inputs = Input((224,224,3))

c1 = Conv2D(128,(3,3),activation='relu',padding='same')(inputs)
c1 = BatchNormalization()(c1)
c1 = Conv2D(128,(3,3),activation='relu',padding='same')(c1)
c1 = BatchNormalization()(c1)
p1 = MaxPooling2D((2,2))(c1)

c2 = Conv2D(64,(3,3),activation='relu',padding='same')(p1)
c2 = BatchNormalization()(c2)
c2 = Conv2D(64,(3,3),activation='relu',padding='same')(c2)
c2 = BatchNormalization()(c2)
p2 = MaxPooling2D((2,2))(c2)

c3 = Conv2D(128,(3,3),activation='relu',padding='same')(p2)
c3 = BatchNormalization()(c3)
c3 = Conv2D(128,(3,3),activation='relu',padding='same')(c3)
c3 = BatchNormalization()(c3)
p3 = MaxPooling2D((2,2))(c3)

c4 = Conv2D(256,(3,3),activation='relu',padding='same')(p3)
c4 = BatchNormalization()(c4)
c4 = Conv2D(256,(3,3),activation='relu',padding='same')(c4)
c4 = BatchNormalization()(c4)
p4 = MaxPooling2D((2,2))(c4)

# c5 = Conv2D(512,(3,3),activation='relu',padding='same')(p4)
# c5 = BatchNormalization()(c5)
# c5 = Conv2D(512,(3,3),activation='relu',padding='same')(c5)
# c5 = BatchNormalization()(c5)
# p5 = MaxPooling2D((2,2))(c5)

c6 = Conv2D(1024,(3,3),activation='relu',padding='same')(p4)
c6 = BatchNormalization()(c6)
c6 = Conv2D(1024,(3,3),activation='relu',padding='same')(c6)
c6 = BatchNormalization()(c6)

# u7 = UpSampling2D((2, 2), interpolation="bilinear")(c6)
# c7 = Concatenate()([u7, c5])
# c7 = Conv2D(512,(3,3),activation='relu',padding='same')(c7)
# c7 = BatchNormalization()(c7)
# c7 = Conv2D(512,(3,3),activation='relu',padding='same')(c7)
# c7 = BatchNormalization()(c7)

u8 = UpSampling2D((2, 2), interpolation="bilinear")(c6)
c8 = Concatenate()([u8, c4])
c8 = Conv2D(256,(3,3),activation='relu',padding='same')(c8)
c8 = BatchNormalization()(c8)
c8 = Conv2D(256,(3,3),activation='relu',padding='same')(c8)
c8 = BatchNormalization()(c8)

u9 = UpSampling2D((2, 2), interpolation="bilinear")(c8)
c9 = Concatenate()([u9, c3])
c9 = Conv2D(128,(3,3),activation='relu',padding='same')(c9)
c9 = BatchNormalization()(c9)
c9 = Conv2D(128,(3,3),activation='relu',padding='same')(c9)
c9 = BatchNormalization()(c9)

u10 = UpSampling2D((2, 2), interpolation="bilinear")(c9)
c10 = Concatenate()([u10, c2])
c10 = Conv2D(64,(3,3),activation='relu',padding='same')(c10)
c10 = BatchNormalization()(c10)
c10 = Conv2D(64,(3,3),activation='relu',padding='same')(c10)
c10 = BatchNormalization()(c10)

u11 = UpSampling2D((2, 2), interpolation="bilinear")(c10)
c11 = Concatenate()([u11, c1])
c11 = Conv2D(128,(3,3),activation='relu',padding='same')(c11)
c11 = BatchNormalization()(c11)
c11 = Conv2D(128,(3,3),activation='tanh',padding='same')(c11)
c11 = BatchNormalization()(c11)

outputs = Conv2D(3, 4, padding="same", activation="softmax")(c11)
model = tf.keras.Model(inputs=[inputs],outputs=[outputs])
```

```python
model.compile(optimizer='adam',loss='mean_squared_error',metrics=['accuracy'])
```

```python
model.fit(train,mask_dataset, batch_size=5, epochs=30)
```

```python
model.save("my_model.h5")
```
